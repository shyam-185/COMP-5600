To get the conditional probability for each feature I first decided the P(A) and P(B) in the equation P(A∣B) = P(A∩B) / P(B)
​P(A) is each feature section (e.g. # Rooms). P(B) is the Construction Type (i.e. Apartment, Condo, House)
So an example would be - P(A∣B) is the probability of 6 Rooms given that it is an House.

for 2.0 Garage we have 2 instances / 7 instances of Garage that is construction type House in our dataset. 2/7 = 0.29

```python
import pandas as pd

conditional_probabilities = {
    'Apartment': {
        'Local Price': {
            (0.0, 2.0): 0.00,
            (2.0, 4.0): 0.00,
            (4.0, 6.0): 0.57,
            (6.0, 8.0): 0.00,
            (8.0, 10.0): 0.29,
            (10.0, 12.0): 0.00,
            (12.0, 14.0): 0.00,
            (14.0, 16.0): 0.14
        },
        'Bathrooms': {
            1.0: 0.71,
            2.5: 0.14,
            1.5: 0.14
        },
        'Land Area': {
            (0.0, 2.0): 0.00,
            (2.0, 4.0): 0.17,
            (4.0, 6.0): 0.67,
            (6.0, 8.0): 0.17,
            (8.0, 10.0): 0.00
        },
        'Living area': {
            (0.0, 2.0): 0.86,
            (2.0, 4.0): 0.14
        },
        '# Garages': {
            1.0: 0.43,
            2.0: 0.29,
            0.0: 0.14,
            1.5: 0.14
        },
        '# Rooms': {
            7: 0.29,
            6: 0.29,
            9: 0.14,
            5: 0.14,
            8: 0.14
        },
        '# Bedrooms': {
            4: 0.29,
            3: 0.43,
            5: 0.14,
            2: 0.14
        },
        'Age of home': {
            (0, 10): 0.00,
            (10, 20): 0.14,
            (20, 30): 0.14,
            (30, 40): 0.00,
            (40, 50): 0.57,
            (50, 60): 0.14
        }
    },
    'House': {
        'Local Price': {
            (0.0, 2.0): 0.00,
            (2.0, 4.0): 0.00,
            (4.0, 6.0): 0.71,
            (6.0, 8.0): 0.29,
            (8.0, 10.0): 0.00,
            (10.0, 12.0): 0.00,
            (12.0, 14.0): 0.00,
            (14.0, 16.0): 0.00
        },
        'Bathrooms': {
            1.0: 0.86,
            1.5: 0.14
        },
        'Land Area': {
            (0.0, 2.0): 0.00,
            (2.0, 4.0): 0.14,
            (4.0, 6.0): 0.29,
            (6.0, 8.0): 0.29,
            (8.0, 10.0): 0.29
        },
        'Living area': {
            (0.0, 2.0): 1.00,
            (2.0, 4.0): 0.00
        },
        '# Garages': {
            2.0: 0.29,
            0.0: 0.29,
            1.0: 0.29,
            1.5: 0.14
        },
        '# Rooms': {
            7: 0.29,
            6: 0.57,
            5: 0.14
        },
        '# Bedrooms': {
            4: 0.14,
            3: 0.71,
            2: 0.14
        },
        'Age of home': {
            (0, 10): 0.00,
            (10, 20): 0.00,
            (20, 30): 0.50,
            (30, 40): 0.50,
        }
    },
    'Condo': {
        'Local Price': {
            (0.0, 2.0): 0.00,
            (2.0, 4.0): 0.20,
            (4.0, 6.0): 0.60,
            (6.0, 8.0): 0.20,
            (8.0, 10.0): 0.00,
            (10.0, 12.0): 0.00,
            (12.0, 14.0): 0.00,
            (14.0, 16.0): 0.00
        },
        'Bathrooms': {
            1.0: 0.67,
            1.5: 0.17,
            2.5: 0.17
        },
        'Land Area': {
            (0.0, 2.0): 0.00,
            (2.0, 4.0): 0.17,
            (4.0, 6.0): 0.33,
            (6.0, 8.0): 0.33,
            (8.0, 10.0): 0.17
        },
        'Living area': {
            (0.0, 2.0): 0.83,
            (2.0, 4.0): 0.17
        },
        '# Garages': {
            1.0: 0.67,
            2.0: 0.33
        },
        '# Rooms': {
            6: 0.67,
            7: 0.17,
            10: 0.17
        },
        '# Bedrooms': {
            3: 0.83,
            5: 0.17
        },
        'Age of home': {
            (0, 10): 0.00,
            (10, 20): 0.17,
            (20, 30): 0.00,
            (30, 40): 0.33,
            (40, 50): 0.17,
            (50, 60): 0.33
        }
    }
}

# Load test data from Excel file
test_data = pd.read_excel("Asssignment2_NB_Data.xlsx", sheet_name = "Test")

# Convert test data to a list of dictionaries
test_data_list = test_data.to_dict(orient = 'records')

# === Prior Probabilities (Calculate these based on your training data) ===
prior_probs = {
    'Apartment': 0.35,
    'House': 0.35,
    'Condo': 0.30
}


def naive_bayes_classifier(test_data, conditional_probabilities, prior_probs):
    classifications = []
    for example in test_data:
        class_probabilities = {}

        for class_type, class_cond_prob in conditional_probabilities.items():
            prob = prior_probs[class_type]
            for feature, value in example.items():
                if feature in class_cond_prob and value in class_cond_prob[feature]:
                    prob *= class_cond_prob[feature][value]
                else:
                    prob *= 0.01  # Smoothing factor for missing values
            class_probabilities[class_type] = prob

        total_prob = sum(class_probabilities.values())
        for class_type in class_probabilities:
            class_probabilities[class_type] /= total_prob

        max_class = max(class_probabilities, key=class_probabilities.get)
        classifications.append((max_class, class_probabilities))

    return classifications


# Classify test data
classifications = naive_bayes_classifier(test_data_list, conditional_probabilities, prior_probs)

# Print the corrected results
for i, (classification, class_probs) in enumerate(classifications):
    print(f"Example {i + 1}:")
    print(f"  - Classified as: {classification}")
    class_probs_rounded = {class_type: round(prob, 2) for class_type, prob in class_probs.items()}
    print(f"  - Class Probabilities: {class_probs_rounded}")
```

```python
import cv2
import numpy as np

def kmeans(img, k, max_iter=100):
    # Flatten image to samples-by-colors shape
    data = img.reshape((-1,3)) 
    data = np.float32(data)

    # Initialize centroids randomly
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, max_iter, 1.0)
    _, labels, centroids = cv2.kmeans(data, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

    # Map pixels back to closest centroids (compressed colors)
    centroids = np.uint8(centroids)
    compressed_img = centroids[labels.flatten()].reshape(img.shape)

    return compressed_img

# Load image
img = cv2.imread('test_image.png')

# Testing usage - compress with k = 50 colors
compressed_img = kmeans(img, k = 50)
cv2.imshow("Compressed Image (k = 50)", compressed_img)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the training and test data
x_train = np.load('x_train.npy')
y_train = np.load('y_train.npy')
x_test = np.load('x_test.npy')
y_test = np.load('y_test.npy')

# Plot the training data
plt.scatter(x_train, y_train, color='blue', label='Training data')
# Plot the test data
plt.scatter(x_test, y_test, color='orange', label='Test data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relationship between x and y')
plt.legend()
plt.show()
```

```python
def error_pred(X, y, predictions):
    m = len(y)
    cost = (1/(2*m)) * np.sum(np.square(predictions - y))
    return cost

def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    error_pred_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = predictions - y 
        theta = theta - (learning_rate / m) * X.transpose().dot(errors)
        error_pred_history[i] = error_pred(X, y, predictions) 

    return theta, error_pred_history

# Add a column of ones to x_train to account for the intercept term (bias)
X_train = np.column_stack((np.ones(len(x_train)), x_train))

# Initialize theta (parameters)
b = np.zeros(X_train.shape[1])

# Set the learning rate and number of iterations
learning_rate = 0.01
iterations = 1000

# Perform gradient descent
b, error_pred_history = gradient_descent(X_train, y_train, b, learning_rate, iterations)

# Plot the regression line
plt.scatter(x_train, y_train, color='blue', label='Training data')
plt.plot(x_train, X_train.dot(b), color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
